```python
%pip install numpy
%pip install pandas
%pip install matplotlib
%pip install yfinance
%pip install scikit-learn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf  #real time financial data of yahoo stock price
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, ElasticNet
from sklearn.svm import SVR
```

```python
google = yf.Ticker("GOOG")
df = google.history(period='1d', interval="1m")
df = df[['Low']]
```

```python
X = np.arange(len(df)).reshape(-1, 1)
y = df['Low'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, shuffle=False)
```

```python
plt.plot(range(0,len(y_train)),y_train, label='Train')
plt.plot(range(len(y_train),len(y)),y_test,label='Test')
plt.legend()
plt.show()
```

```python
# Create a linear regression model
linear_model = LinearRegression()
lasso_model = Lasso()
elasticnet_model = ElasticNet()
svr_model = SVR()
```

```python
linear_model.fit(X_train, y_train)
lasso_model.fit(X_train, y_train)
elasticnet_model.fit(X_train, y_train)
svr_model.fit(X_train, y_train)
```

```python
score = linear_model.score(X_test, y_test)
print("linear_model Model score:", score)
score = lasso_model.score(X_test, y_test)
print("lasso_model Model score:", score)
score = elasticnet_model.score(X_test, y_test)
print("elasticnet_model Model score:", score)
score = svr_model.score(X_test, y_test)
print("svr_model Model score:", score)
Linear_predictions = linear_model.predict(X_test)
lasso_predictions = lasso_model.predict(X_test)
elasticnet_predictions = elasticnet_model.predict(X_test)
svr_predictions = svr_model.predict(X_test)
```

```python
plt.plot(X_train, y_train, label='Train')
plt.plot(X_test, y_test, label='Test')
plt.plot(X_test, Linear_predictions, label='Linear Regression Predictions')
plt.plot(X_test, lasso_predictions, label='Lasso Predictions')
plt.plot(X_test, elasticnet_predictions, label='ElasticNet Predictions')
plt.plot(X_test, svr_predictions, label='SVR Predictions')

plt.legend()
plt.show()
```
